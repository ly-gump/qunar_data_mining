```python
import requests
import lxml.html
import codecs
import time
import re
import pandas as pd
```

```python
def get_text_from_qunar(url):
    myheaders = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)' +  
                 'Chrome/94.0.4606.61 Safari/537.36'}
    http_response = requests.get(url, headers=myheaders)
    http_response.encoding = 'utf-8'

    html = lxml.html.fromstring(http_response.text)
    info = html.xpath('//p')
    text = ''
    for line in info:
        text += str(line.text_content())
    return text
```

```python
url = 'https://travel.qunar.com/travelbook/note/7407474'
%timeit print(get_text_from_qunar(url))
```

```python
qunar = pd.read_csv('qunar_data.csv')
```

```python
data = {}
count = 0
for link in qunar['链接']:
    data.update({count:get_text_from_qunar(link)})
    count += 1
qunar_text = pd.Series(data)
```
